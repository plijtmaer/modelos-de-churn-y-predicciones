#### Hacemos fit con los datos, predecimos y evaluamos m√©tricas

```python
# Fitteamos
knn.fit(X_train, y_train)

# Predecimos etiquetas para los datos de test
y_pred_knn = knn.predict(X_test)

# Predecimos probabilidades
y_pred_proba_knn = knn.predict_proba(X_test)

# Creamos variables para accuracy, recall, precision, f1-score
knn_accuracy = accuracy_score(y_test, y_pred_knn)
knn_recall = recall_score(y_test, y_pred_knn)
knn_precision = precision_score(y_test, y_pred_knn)
knn_f1 = f1_score(y_test, y_pred_knn)
```

```python
# Evaluamos la curva ROC y el AUC
fpr_log_knn, tpr_log_knn, thr_log_knn = roc_curve(y_test, y_pred_proba_knn[:,1])

# Creamos un DataFrame
knn_roc_df = pd.DataFrame(dict(fpr=fpr_log_knn, tpr=tpr_log_knn, thr=thr_log_knn))

# Visualizamos la curva ROC
plt.figure(figsize=(4,3))
plt.axis([0, 1.01, 0, 1.01])
plt.xlabel('1 - Specificity')
plt.ylabel('TPR / Recall')
plt.title('ROC Curve')
plt.plot(knn_roc_df['fpr'], knn_roc_df['tpr'])
plt.plot(np.arange(0,1, step=0.01), np.arange(0,1, step=0.01))

# Creamos una variable para el AUC
knn_auc = auc(fpr_log_knn, tpr_log_knn)
print(f'AUC: {knn_auc}')
```

## 1.2 KNN con distintos n_neighbors

#### Vamos a evaluar KNN con distintos n_neighbors, adem√°s del default de 5 vecinos, para ver si mejora el modelo

```python
# Vamos a trabajar sobre los datos de train, ya que no podemos evaluar accuracy sobre los datos de test porque estamos en la etapa de elegir los mejores hiperpar√°metros para el modelo
X_train_train_knn, X_validation_knn, y_train_train_knn, y_validation_knn = train_test_split(X_train, y_train, random_state=7)
```

```python
# Creamos una lista para almacenar el accuracy score para cada n_neighbors entre 1 y 25.
scores = []

# Creamos un loop donde instanciamos knn con la distinta cantidad de vecinos, luego fit, predecimos y agregamos los puntajes a la lista
for k in range(1, 26):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_train_knn, y_train_train_knn)
    y_pred = knn.predict(X_validation_knn)
    scores.append(accuracy_score(y_validation_knn, y_pred))
```

```python
# Visualizamos los k vecinos contra sus respectivos valores de accuracy
plt.figure(figsize=(4,3))
plt.plot(range(1, 26), scores)
plt.xlabel('K - Neighbors')
plt.ylabel('Accuracy Score')
plt.grid()
```

```python
# El accuracy m√°s alto est√° aproximadamente entre el valor 7 y el 8 pero tenemos que encontrar el valor exacto
l =  [i for i, val in enumerate(scores) if val==max(scores)]

# Le sumamos 1 porque arrancamos en 1 neighbor
print(max(l) + 1)
```

#### Al encontrar que con 8 vecinos KNN tiene una mejor accuracy, vamos a instanciarlo y computar todas las m√©tricas

```python
# Fitteamos
knn_n = KNeighborsClassifier(n_neighbors=8)
knn_n.fit(X_train, y_train)

# Predecimos etiquetas para los datos de test
y_pred_knn_n = knn_n.predict(X_test)

# Predecimos probabilidades
y_pred_proba_knn_n = knn_n.predict_proba(X_test)

# Creamos variables para accuracy, recall, precision, f1-score
knn_n_accuracy = accuracy_score(y_test, y_pred_knn_n)
knn_n_recall = recall_score(y_test, y_pred_knn_n)
knn_n_precision = precision_score(y_test, y_pred_knn_n)
knn_n_f1 = f1_score(y_test, y_pred_knn_n)
```

```python
# Evaluamos la curva ROC y el AUC
fpr_log_knn_n, tpr_log_knn_n, thr_log_knn_n = roc_curve(y_test, y_pred_proba_knn_n[:,1])

# Creamos un DataFrame
knn_n_roc_df = pd.DataFrame(dict(fpr=fpr_log_knn_n, tpr=tpr_log_knn_n, thr=thr_log_knn_n))

# Visualizamos la curva ROC
plt.figure(figsize=(4,3))
plt.axis([0, 1.01, 0, 1.01])
plt.xlabel('1 - Specificity')
plt.ylabel('TPR / Recall')
plt.title('ROC Curve')
plt.plot(knn_n_roc_df['fpr'], knn_n_roc_df['tpr'])
plt.plot(np.arange(0,1, step=0.01), np.arange(0,1, step=0.01))

# Creamos una variable para el AUC
knn_n_auc = auc(fpr_log_knn_n, tpr_log_knn_n)
print(f'AUC: {knn_n_auc}')
```

```python
# Evaluamos la matriz de confusi√≥n
plt.figure(figsize=(4,3))
knn_n_cm = confusion_matrix(y_test, y_pred_knn_n)
sns.heatmap(knn_n_cm, annot=True, xticklabels=True, yticklabels=True, fmt='.4g')
```

```python
# Evaluamos todas las m√©tricas para este KNN simple
knn_df = pd.DataFrame({'Accuracy': [knn_accuracy, knn_n_accuracy], 'Recall': [knn_recall, knn_n_recall], 'Precision': [knn_precision, knn_n_precision],\
                      'F1-Score': [knn_f1, knn_n_f1], 'AUC': [knn_auc, knn_n_auc]}, index=['KNN B√°sico', 'KNN Diferentes Vecinos'])
knn_df
```

#### Observamos que el accuracy aument√≥, junto a la presicion y al AUC, mientras que recall y F1-score disminuyeron

## 1.3 KNN con Cross-Validation

#### Vamos a aplicar cross-validation para ver c√≥mo mejora KNN y si es m√°s exacto

```python
# Usamos cross validation con 5 folds
kf = KFold(n_splits=5, shuffle=True, random_state=7)

# Creamos una lista para almacenar los scores
knn_cv_scores_para_df = []

# Creamos un loop donde instanciamos knn con la distinta cantidad de vecinos, luego fit, predecimos y agregamos los puntajes a la lista
for i in range(1, 26):

    # En cada iteraci√≥n, instanciamos KNN con un hiperpar√°metro distinto
    knn = KNeighborsClassifier(n_neighbors=i)

    # cross_val_scores devuelve un array de 5 resultados, uno por cada fold que hizo autom√°ticamente CV
    cv_scores = cross_val_score(knn, X_train, y_train, cv=kf)

    # Para cada valor del hiperpar√°metro, creamos un diccionario con el valor el valor de vecinos y la media de los scores
    dict_row_score = {'score_medio': np.mean(cv_scores), 'n_neighbors': i}

    # Almacenamos en la lista de diccionarios
    knn_cv_scores_para_df.append(dict_row_score)

# Creamos un DataFrame a partir de la lista de diccionarios
knn_cv_df_scores = pd.DataFrame(knn_cv_scores_para_df)
```

```python
# Visualizamos el DataFrame creado
knn_cv_df_scores.sample(3)
```

```python
# Visualizamos la media por cada valor de n_neighbors
plt.figure(figsize=(4,3))
plt.plot(knn_cv_df_scores['n_neighbors'], knn_cv_df_scores['score_medio'])
plt.xlabel('K - Neighbors')
plt.ylabel('Accuracy')
plt.grid()
```

```python
# Encontramos el n√∫mero de n_neighbors para el score medio
knn_cv_df_scores[knn_cv_df_scores['score_medio'] == knn_cv_df_scores['score_medio'].max()]
```

#### Al encontrar que hay 2 n√∫meros de vecinos con el mismo score_medio m√°ximo, nos quedamos con el valor menor de vecinos, 9.
#### Instanciamos KNN, computamos las m√©tricas y comparamos

```python
# Fitteamos
knn_cv = KNeighborsClassifier(n_neighbors=9)
knn_cv.fit(X_train, y_train)

# Predecimos etiquetas para los datos de test
y_pred_knn_cv = knn_cv.predict(X_test)

# Predecimos probabilidades
y_pred_proba_knn_cv = knn_cv.predict_proba(X_test)

# Creamos variables para accuracy, recall, precision, f1-score
knn_cv_accuracy = accuracy_score(y_test, y_pred_knn_cv)
knn_cv_recall = recall_score(y_test, y_pred_knn_cv)
knn_cv_precision = precision_score(y_test, y_pred_knn_cv)
knn_cv_f1 = f1_score(y_test, y_pred_knn_cv)
```

```python
# Evaluamos la curva ROC y el AUC
fpr_log_knn_cv, tpr_log_knn_cv, thr_log_knn_cv = roc_curve(y_test, y_pred_proba_knn_cv[:,1])

# Creamos un DataFrame
knn_cv_roc_df = pd.DataFrame(dict(fpr=fpr_log_knn_cv, tpr=tpr_log_knn_cv, thr=thr_log_knn_cv))

# Visualizamos la curva ROC
plt.figure(figsize=(4,3))
plt.axis([0, 1.01, 0, 1.01])
plt.xlabel('1 - Specificity')
plt.ylabel('TPR / Recall')
plt.title('ROC Curve')
plt.plot(knn_cv_roc_df['fpr'], knn_cv_roc_df['tpr'])
plt.plot(np.arange(0,1, step=0.01), np.arange(0,1, step=0.01))

# Creamos una variable para el AUC
knn_cv_auc = auc(fpr_log_knn_cv, tpr_log_knn_cv)
print(f'AUC: {knn_cv_auc}')
```

```python
# Evaluamos la matriz de confusi√≥n
plt.figure(figsize=(4,3))
knn_cv_cm = confusion_matrix(y_test, y_pred_knn_cv)
sns.heatmap(knn_cv_cm, annot=True, xticklabels=True, yticklabels=True, fmt='.4g')
```

```python
# Evaluamos todas las m√©tricas
knn_df = pd.DataFrame({'Accuracy': [knn_accuracy, knn_n_accuracy, knn_cv_accuracy], \
                       'Recall': [knn_recall, knn_n_recall, knn_cv_recall],\
                       'Precision': [knn_precision, knn_n_precision, knn_cv_precision],\
                       'F1-Score': [knn_f1, knn_n_f1, knn_cv_f1],\
                       'AUC': [knn_auc, knn_n_auc, knn_cv_auc]},\
                      index=['KNN B√°sico', 'KNN Diferentes Vecinos', 'KNN CV'])
knn_df
```

## 1.4 KNN con RandomizedSearchCV

#### Vamos a seleccionar de forma aleatoria un subset de los hiperpar√°metros con RandomizedSearchCV

```python
# Evaluamos los resultados generados
pd.DataFrame(rs.cv_results_).head(3)
```

```python
# Buscamos los mejores hiperpar√°metros
print(rs.best_estimator_)
print(rs.best_score_)
print(rs.best_params_)
```

## 1.6 KNN con Distintas Probabilidades

#### Luego de obtener el mejor KNN tras 5 instancias de algoritmos distintos, vamos a probar cambiar la probabilidad y evaluar si hay alguna mejora

```python
# Creamos una lista de umbrales para utilizar para probar distintas probabilidades y su accuracy
thresh_knn = np.arange(0.1, 1, 0.05)

# Instanciamos una lista para almacenar los outputs del loop
thresh_knn_list = []

# Armamos un loop con distintas probabilidades para ver si mejora la accuracy
for thresh in thresh_knn:

    # Creamos un nuevo y_pred utilizando la funci√≥n binarize para cambiar el threshold de la probabilidad
    y_pred_thresh_knn = binarize(y_pred_proba_knn_gs, threshold=thresh)[:,1]

    # Creamos un diccionario para almacenar resultados
    dict_row_score = {'Accuracy Score': accuracy_score(y_test, y_pred_thresh_knn), 'Probabilidad': thresh}

    # Agregamos el diccionario a la lista
    thresh_knn_list.append(dict_row_score)
```

```python
# Creamos un DataFrame para visualizar los resultados
thresh_knn_df = pd.DataFrame(thresh_knn_list)

# Visualizamos el accuracy score m√°s alto y su probabilidad
thresh_knn_max = thresh_knn_df['Accuracy Score'].max()
thresh_knn_df[thresh_knn_df['Accuracy Score'] == thresh_knn_max]
```

```python
# Al tener 3 distintas probabilidades, vamos a evaluar m√°s m√©tricas antes de decidir, y debido a que en un modelo de churn bancario, la retenci√≥n de clientes es importante, vamos a darle mayor prioridad al recall sobre la precisi√≥n

# Es fundamental identificar correctamente a los clientes que tienen la intenci√≥n de abandonar el banco (minimizar los falsos negativos), incluso si esto genera un aumento en los falsos positivos.

# Creamos una lista de umbrales con las probabilidades obtenidas en el paso anterior
thresh_knn_2 = [0.4, 0.45, 0.5]

# Instanciamos una lista para almacenar los outputs del loop
thresh_knn_list_2 = []

# Armamos un loop con las distintas probabilidades para comparar accuracy y recall
for thresh in thresh_knn_2:

    # Creamos un nuevo y_pred utilizando la funci√≥n binarize para cambiar el threshold de la probabilidad
    y_pred_thresh_knn = binarize(y_pred_proba_knn_gs, threshold=thresh)[:,1]

    # Creamos un diccionario para almacenar resultados
    dict_row_score = {'Accuracy Score': accuracy_score(y_test, y_pred_thresh_knn),
                      'Recall': recall_score(y_test,y_pred_thresh_knn),
                      'Probabilidad': thresh}

    # Agregamos el diccionario a la lista
    thresh_knn_list_2.append(dict_row_score)
```

```python
# Creamos un DataFrame para visualizar los resultados
thresh_knn_df_2 = pd.DataFrame(thresh_knn_list_2)
thresh_knn_df_2
```

```python
# Encontramos que el threshold de 0.4 es el que tiene el recall m√°s alto, por lo tanto, lo usaremos para computar el resto de las m√©tricas
y_pred_knn_thresh = binarize(y_pred_proba_knn_gs, threshold=0.4)[:,1]
y_pred_proba_knn_thresh = binarize(y_pred_proba_knn_gs, threshold=0.4)

# Creamos variables para accuracy, recall, precision, f1-score
knn_thresh_accuracy = accuracy_score(y_test, y_pred_knn_thresh)
knn_thresh_recall = recall_score(y_test, y_pred_knn_thresh)
knn_thresh_precision = precision_score(y_test, y_pred_knn_thresh)
knn_thresh_f1 = f1_score(y_test, y_pred_knn_thresh)
```

```python
# Generamos la variables predictoras y variable objetivo
X = df.loc[:, df.columns != 'churn']
y = df.loc[:,'churn']
```

```python
def train_test_scaled(X,y):
    """Funci√≥n para generar los dataset de entrenamiento y testeo y luego escalarlos"""
    # Hacemos el split entre train y test
    global y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=y, random_state=7)
    # Utilizamos sklearn para estandarizar la matriz de Features
    scaler = StandardScaler()
    global X_train_scaled
    X_train_scaled = scaler.fit_transform(X_train)
    # Escalamos el dataset de testeo
    global X_test_scaled
    X_test_scaled = scaler.transform(X_test)
```

```python
# Aplicamos la funcion a X e y
train_test_scaled(X,y)
```

```python
# Funcion con statsmodels
def reglog_stats(X,y):
    """Funcion que genera un modelo de reg logistica con statsmodels
    y tambi√©n imprime las columnas con p values mayor a 5%"""
    # Agregamos la constante beta cero
    X_train_stats = sm.add_constant(X)
    # Instanciamos la clase
    logit = sm.Logit(y, X_train_stats)
    # Fiteamos el modelo
    global result
    result = logit.fit()
    # Imprimimos el resumen
    print(result.summary())
    # Imprimimos valores con pvalue menor al 5%
    global df_pvalues # To create a global variable inside a function
    df_pvalues = pd.DataFrame(result.pvalues[result.pvalues > 5/100], columns=['valor']).reset_index()
```

```python
reglog_stats(X_train_scaled,y_train)
```

```python
X_test_scaled.shape
```

La conclusion es que la probabilidad de darse de baja no depende del `credit_card`, `estimated_salary`, `country_Spain`. Dado que el valor p asociado a estas features en la tabla anterior es mayor a 0.05, no podemos rechazar $H_0$.

```python
# Identificamos la variables con pvalue mayor a 0.05
var_num = np.array(df_pvalues['index'].str.extract('(\d+)').astype(int)).flatten()
print(f'Variables eliminadas: {X.columns[var_num-1].tolist()}')
# Eliminamos las variables con pvalue mayor a 0.05
X_nuevo = X.drop(X.columns[var_num-1], axis=1)
print(f'Features nuevas: {X_nuevo.columns.tolist()}')
```

```python
# Aplicamos funcion para escalar
train_test_scaled(X_nuevo,y)
# Aplicamos funcion para realizar la reg log
reglog_stats(X_train_scaled,y_train)
```

Todos los p-values son mayores a 0.05, concluimos que s√≠ existe una asociaci√≥n entre `churn` y 'credit_score', 'age', 'tenure', 'balance', 'products_number', 'active_member', 'country_Germany', 'gender_Male'.

## 3.3 M√©tricas

```python
# Elaboramos la matriz de confusi√≥n
def cm(y_test, y_pred):
    confusion_matrix(y_test, y_pred)
    xlabel=['Predicho : NO', 'Predicho : SI']
    ylabel=['Verdadero : NO', 'Verdadero : SI']
    print("CONFUSION MATRIX")
    sns.heatmap(confusion_matrix(y_test, y_pred),
                                annot=True,
                                fmt='g',
                                xticklabels=xlabel,
                                yticklabels=ylabel)

```

```python
# Funcion para obtener m√©tricas
def metrics_dict(y_test,y_pred):
    dict = {}
    dict['accuracy'] = accuracy_score(y_test, y_pred).round(2)
    dict['recall'] = recall_score(y_test, y_pred).round(2)
    dict['precision'] = precision_score(y_test, y_pred).round(2)
    dict['f1_score'] = f1_score(y_test, y_pred).round(2)
    return dict
```

```python
metrics_1 = metrics_dict(y_test, y_pred_1)
metrics_2 = metrics_dict(y_test, y_pred_2)
metrics_3 = metrics_dict(y_test, y_pred_3)
metrics_4 = metrics_dict(y_test, y_pred_4)
metrics_5 = metrics_dict(y_test, y_pred_5)
```

```python
resultados = pd.DataFrame([metrics_1, metrics_2, metrics_3, metrics_4, metrics_5])
resultados.index = ['caso_1', 'caso_2', 'caso_3', 'caso_4', 'caso 5']
```

Curva ROC y AUC

```python
# Funcion que genera curva ROC y AUC
def roc_auc(y_test, probs):
    fpr,tpr,_ = roc_curve(y_test, probs[:,1])
    dft = pd.DataFrame(dict(fpr=fpr, tpr=tpr))
    plt.axis([0, 1.01, 0, 1.01])
    plt.xlabel('1 - Specificty')
    plt.ylabel('TPR / Sensitivity')
    plt.title('ROC Curve')
    plt.plot(dft['fpr'],dft['tpr'])
    plt.plot(np.arange(0,1, step =0.01), np.arange(0,1, step =0.01))
    print('√ÅREA BAJO LA CURVA')
    print('AUC: ',auc(fpr,tpr))
    print('----------------')
    print('CURVA ROC')
```
